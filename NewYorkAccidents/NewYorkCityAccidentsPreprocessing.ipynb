We have a dataset that includes traffic accidents that occurred in New York City. This dataset includes motor vehicle accidents reported between January and August 2020. This study is being conducted in order to be aware of and take precautions against accidents that may occur in New York City in the future.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# While preparing the dataset, we call the libraries that will be useful to us.
```

```python
veri=pd.read_csv("Accidents")
#I call my dataset under the variable name 'veri'.
```

```python
# I select the rows with fatal accidents to visualize the number and types of deaths.
killed_columns = ['NUMBER OF PERSONS KILLED','NUMBER OF PEDESTRIANS KILLED', 'NUMBER OF CYCLIST KILLED', 'NUMBER OF MOTORIST KILLED']

# I create a new df from the selected rows.
killed_data = veri[killed_columns]
# I am organizing it according to Seaborn
killed_data_melted = killed_data.melt(value_vars=killed_columns, var_name='Person Type', value_name='Number of Killed Persons')

# I am creating a graphic
plt.figure(figsize=(10, 6))
sns.countplot(x='Person Type', hue='Number of Killed Persons', data=killed_data_melted)
plt.title('Number of Killed Persons by Person Type')
plt.xlabel('Person Type')
plt.ylabel('Count')
plt.show()
```

```python
# For clearer readability, I create a new df that contains only fatal accidents and visualize it.
verif1 = veri[veri['NUMBER OF PERSONS KILLED'] > 0].copy()
```

```python
killed_data_verif1 = verif1[killed_columns]

killed_data_verif1_melted = killed_data_verif1.melt(value_vars=killed_columns, var_name='Person Type', value_name='Number of Killed Persons')

plt.figure(figsize=(10, 6))
sns.countplot(x='Person Type', hue='Number of Killed Persons', data=killed_data_verif1_melted)
plt.title('Number of Killed Persons in verif1 by Person Type')
plt.xlabel('Person Type')
plt.ylabel('Count')
plt.show()
```

```python
# I group accidents according to districts.
plt.figure(figsize=(12, 6))
sns.countplot(x='BOROUGH', data=verif1, order=verif1['BOROUGH'].value_counts().index)
plt.title('Number of Accidents in Each Borough (verif1)')
plt.xlabel('Borough')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# To handle more serious accidents, I create a new df for accidents with more than 3 casualties.
verif2 = veri[veri['NUMBER OF PERSONS INJURED'] > 3].copy()
```

Before performing Encoding on the currently available version of the Data Set, visualization studies will be performed to check whether there is inconsistent data or incompatible content.

```python
numerical_cols = ['NUMBER OF PERSONS INJURED', 'NUMBER OF PERSONS KILLED', 'NUMBER OF PEDESTRIANS INJURED', 'NUMBER OF PEDESTRIANS KILLED', 'NUMBER OF CYCLIST INJURED', 'NUMBER OF CYCLIST KILLED', 'NUMBER OF MOTORIST INJURED', 'NUMBER OF MOTORIST KILLED']

correlation_matrix = vericnd[numerical_cols].corr()

plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Heatmap')
plt.show()
```

In this graph, it is seen that motorcycle drivers constitute the most injury accidents with a weight of 0.98. At the same time, pedestrians and cyclists are less involved in injury accidents with a weight of 0.05.

In fatal accidents, motorcycle drivers have the highest value with 0.72, while pedestrians are in second place with 0.58. In fatal accidents, the least loss of life is in cyclists.
